```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit



def eV2au(E):
  return E/27.211

def nm2au(x):
  return x*18.897259886


l = nm2au(10)
a = nm2au(30)
epsilon_r = 12.5
kappa = 1/epsilon_r
m = 0.067
n = 41
dx = 2*a/(n-1)
dt = m*dx**2*0.4
tol = 10**-9

def print_matrix(M):

  for i in range(len(M)):
    for j in range(len(M[0])):
      print('{:<10.3}'.format(M[i,j]), end=' ')
    print()

@njit
def Vint(x1, x2):
  return kappa / np.sqrt((x1-x2)**2 + l**2)


###### normowanie funckcji ######
@njit
def norm(psi,dx):

  c = 0
  if psi.ndim == 2:
    for i in range(len(psi)):
      for j in range(len(psi[0])):
        c += abs(psi[i,j])**2 * dx**2

  elif psi.ndim == 1:
    for i in range(len(psi)):
      c += abs(psi[i])**2 * dx**2
  return psi / np.sqrt(c)

###### obliczanie energii ######
@njit
def energy(psi,dx,n,a):

  E = 0
  f_arr = f(psi,n,a)
  for i in range(len(psi)):
    for j in range(len(psi[0])):
      E += psi[i,j] * f_arr[i,j] * dx**2
  return E

###### obliczanie H*psi ######
@njit
def f(psi,n,a):

  f_arr = np.zeros((n,n))
  dx = 2*a/(n-1)

  for i in range(1,n-1):
    for j in range(1,n-1):
      x1 = -a+i*dx
      x2 = -a+j*dx
      f_arr[i,j] = -(psi[i+1,j] + psi[i-1,j] + psi[i,j+1] + psi[i,j-1] - 4*psi[i,j]) / (dx**2 * 2*m) + Vint(x1,x2)*psi[i,j]
  f_arr[0,:] = f_arr[n-1,:] = f_arr[:,0] = f_arr[:,n-1] = 0
  return f_arr

###### inicjacja psi wartościami losowymi ######
@njit
def make_psi(n):
    psi = np.empty((n, n), dtype=np.float64)
    for i in range(n):
        for j in range(n):
            psi[i, j] = 2 * (np.random.random() - 0.5)
    psi[0, :] = psi[n-1, :] = psi[:, 0] = psi[:, n-1] = 0
    return psi

###### algorytm czasu urojonego ######
@njit
def IT(n,a):

  dx = 2*a/(n-1)
  dt = m*dx**2*0.4
  counter = 0
  E_arr = []

  Psi = make_psi(n)

  condition = False
  while condition == False:

    Ecurr = energy(Psi,dx,n,a)
    Psi_next = norm(Psi - dt * f(Psi,n,a), dx)
    Enext = energy(Psi_next,dx,n,a)

    if abs((Enext - Ecurr) / Enext) < tol:
      condition = True
    else:
      Psi = Psi_next

    E_arr.append(Enext)
    counter += 1

  return E_arr, Psi

########################################################################################
################################## Hartree-Fock ########################################

###### inicjacja psi(0) ######
@njit
def make_psi_hf(n):
    psi = np.empty((n), dtype=np.float64)
    for i in range(n):
      psi[i] = 2 * (np.random.random() - 0.5)
    psi[0] = psi[n-1] = 0
    return psi


@njit
def hpsi(psi,i,dx):
  return -(psi[i+1] + psi[i-1] - 2*psi[i]) / (2*m * dx**2)

@njit
def V(psi,i,a,dx):
  J = 0
  for j in range(1,n-1):
    x1 = -a+i*dx
    x2 = -a+j*dx
    J+= abs(psi[j]**2)*Vint(x1,x2)*dx
  return J*psi[i]

@njit
def F_hf(psi,n,dx):
  F = np.zeros(n)
  for i in range(1,n-1):
    F[i] = (hpsi(psi,i,dx) + V(psi,i,a,dx))
  F[0] = F[n-1] = 0
  return F

###### energia metodą czasu urojonego przy F = h + V ######
@njit
def energy_hf(psi,dx,n):
  E = 0
  f_arr = F_hf(psi,n,dx)
  for i in range(len(psi)):
    E += psi[i] * f_arr[i] * dx**2
  return E


@njit
def Eint(psi,a,dx):
  E = 0
  for i in range(n):
    E += abs(psi[i]**2) * V(psi,i,a,dx)
  return E

###### algorytm Hartree-Focka ######
def HF(n,a):
  dx = 2*a/(n-1)
  dt = m*dx**2*0.4
  counter = 0
  E_arr = []
  Eint_arr = []
  Psi = make_psi_hf(n)

  condition = False
  while condition == False:

    # obliczanie E(k)
    Ecurr = energy_hf(Psi,dx,n)
    Eint_curr = Eint(Psi,a,dx)
    Etot_curr = 2*Ecurr - Eint_curr

    Psi_next = norm(Psi - dt * F_hf(Psi,n,dx), dx)

    # obliczanie E(k+1)
    Enext = energy_hf(Psi_next,dx,n)
    Eint_next = Eint(Psi_next,a,dx)
    Etot_next = 2*Enext - Eint_next

    if abs((Etot_next - Etot_curr) / Etot_next) < tol:
      condition = True
    else:
      Psi = Psi_next

    E_arr.append(Etot_next)
    counter += 1

  return E_arr, Psi
```

```python
############################################
################# zad 1 ####################
E, Psi = IT(n,a)

plt.ylabel("E[eV]")
plt.xlabel("iter")
# plt.yscale('log')
plt.plot(np.array(E)/eV2au(1), 'k.', markersize=2);
plt.show()
```

```python
############################################
################# zad 2 ####################

alist = np.arange(30,65,5)
E_list_IT = []
E_list_HF = []
for a in alist:
  E_list_IT.append(IT(n,nm2au(a))[0][-1]/eV2au(1))
  E_list_HF.append(HF(n,nm2au(a))[0][-1]/eV2au(1))

plt.ylabel("E[eV]")
plt.xlabel("a[nm]")
plt.plot(alist, E_list_IT, 'm.');
y_interpolated_it = np.interp(alist, alist, E_list_IT);
plt.xlabel("a [nm]")
plt.plot(alist, y_interpolated_it, label='imaginary time', color = 'purple');

plt.plot(alist, E_list_HF, 'c.');
y_interpolated_hf = np.interp(alist, alist, E_list_HF);
plt.plot(alist, y_interpolated_hf, label='Hartree-Fock', color ='cyan');
plt.legend()
plt.show()
```
