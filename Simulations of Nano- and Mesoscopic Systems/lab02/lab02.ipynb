```python
import numpy as np
import matplotlib.pyplot as plt
import random
from numba import njit
```

```python
Ha = 27.211 # 1 Hartree in eV
a0 = 0.05292 # Bohr radius in nm
m = 0.067 # effective mass for GaAs
eps = 12.5 
n = 41
l = 10 / a0
tol = 1e-9
kappa = 1 / eps
a = 30 / a0
```

```python
@njit
def norm_1d(psi,dx):
    c = 0
    for i in range(len(psi)):
        c += abs(psi[i])**2 * dx**2
    normed_psi = psi / np.sqrt(c)
    return normed_psi

@njit
def norm_2d(psi,dx):
    c = 0
    for i in range(len(psi)):
        for j in range(len(psi[0])):
            c += abs(psi[i][j])**2 * dx**2
    normed_psi = psi / np.sqrt(c)
    return normed_psi
```

## Zadanie 1

```python
@njit
def V_int(x1, x2):
    return kappa / np.sqrt((x1 - x2)**2 + l**2)

@njit
def Psi_2d(n):
    psi = np.zeros((n, n))
    for i in range(1, n - 1):
        for j in range(1, n - 1):
            psi[i, j] = random.uniform(-1, 1)
    return psi

@njit
def fun(psi, n, a):
    dx = 2 * a / (n - 1)
    dt = m * dx * dx * 0.4
    X1 = np.arange(-a, a + dx, dx)
    X2 = np.arange(-a, a + dx, dx)
    f = np.zeros((n, n))
    for i in range(1, n - 1):
        for j in range(1, n - 1):
            f[i][j] = - (psi[i + 1,j] + psi[i - 1,j] + psi[i,j + 1] + psi[i,j - 1] - 4 * psi[i,j]) / (2 * m * dx * dx) + V_int(X1[i],X2[j]) * psi[i,j]
    return f

@njit
def energy(psi, n, a):
    E = 0
    f = fun(psi, n, a)
    dx = 2 * a / (n - 1)
    for i in range(1,n-1):
        for j in range(1,n-1):
            E += psi[i][j] * f[i][j] * dx * dx
    return E
```

```python
@njit
def ImaginaryTime(n, a):
    PSI = Psi_2d(n)
    Energy = []
    dx = 2 * a / (n - 1)
    dt = m * dx * dx * 0.4
    X1 = np.arange(-a, a + dx, dx)
    X2 = np.arange(-a, a + dx, dx)
    while True:
        E_curr = energy(PSI, n, a)
        PSI_next = norm_2d(PSI - dt * fun(PSI, n, a),dx)
        E_next = energy(PSI_next, n, a)
        Energy.append(E_next)
        if abs((E_next - E_curr) / E_next) < tol:
            break
        else:
            PSI = PSI_next
            
    return PSI, Energy
```

```python
PSI, Energy = ImaginaryTime(n, a)
Energy_ha = np.array(Energy) * Ha
```

```python
plt.figure(figsize=(10,6))
plt.ylabel("Energy [eV]")
plt.xlabel("n")

# Wybierz co któryś punkt do wyświetlenia (krok 100)
selected_indices = np.arange(len(Energy_ha))[::100]
selected_energy = Energy_ha[::100]

# Pomijamy pierwszy punkt
selected_indices = selected_indices[1:]
selected_energy = selected_energy[1:]

# plt.plot(selected_indices, selected_energy, 'k.-', markersize=10)
plt.plot(np.arange(len(Energy_ha))[::50], Energy_ha[::50], 'k.', markersize=6)

plt.yticks(np.arange(0.01, 0.02, 0.001))
plt.ylim(0.01, 0.02)

plt.show()
```

## Hartree-Fock

```python
@njit
def Psi_1d(n):
    psi = np.zeros(n)
    for i in range(1, n - 1):
        psi[i] = np.random.uniform(-1, 1)
    return psi

@njit
def hpsi(psi, i, dx):
    return -(psi[i+1] + psi[i-1] - 2 * psi[i]) / (2 * m * dx**2)

@njit
def V(psi, i, a):
    J = 0
    dx = 2 * a / (n - 1)
    dt = m * dx * dx * 0.4
    X1 = np.arange(-a, a + dx, dx)
    X2 = np.arange(-a, a + dx, dx)
    for j in range(1, n-1):
        J += np.abs(psi[j]**2) * V_int(X1[i], X2[j]) * dx
    return J * psi[i]

@njit
def F_hf(psi, n, a):
    F = np.zeros(n)
    dx = 2 * a / (n - 1)
    for i in range(1, n-1):
        F[i] = hpsi(psi, i, dx) + V(psi, i, a)
    return F

@njit
def energy_hf(psi, n, a):
    E = 0
    f_arr = F_hf(psi, n, a)
    dx = 2 * a / (n - 1)
    for i in range(1, n-1):
        E += psi[i] * f_arr[i] * dx**2
    return E

@njit
def Eint(psi, a):
    E = 0
    for i in range(1, n-1):
        E += np.abs(psi[i]**2) * V(psi, i, a)
    return E
```

```python
@njit
def HartreeFock(n, a):
    Energy = []
    Psi = Psi_1d(n)
    dx = 2 * a / (n - 1)
    dt = m * dx * dx * 0.4
    X1 = np.arange(-a, a + dx, dx)
    X2 = np.arange(-a, a + dx, dx)
    
    while True:
        Ecurr = energy_hf(Psi, n, a)
        Eint_curr = Eint(Psi, a)
        Etot_curr = 2 * Ecurr - Eint_curr

        Psi_next = norm_1d(Psi - dt * F_hf(Psi, n, a), dx)

        Enext = energy_hf(Psi_next, n, a)
        Eint_next = Eint(Psi_next, a)
        Etot_next = 2 * Enext - Eint_next

        if abs((Etot_next - Etot_curr) / Etot_next) < tol:
            break
        else:
            Psi = Psi_next
            
        Energy.append(Etot_next)

    return Psi, Energy
```

```python
alist = np.arange(30, 65, 5)
E_list_IT = []
E_list_HF = []
```

```python
for i in range(0, np.size(alist)):
    a = alist[i] / a0
    E_list_IT.append((ImaginaryTime(n, a)[1])[-1] * Ha)
    E_list_HF.append((HartreeFock(n, a)[1])[-1] * Ha)
```

## Zadanie 2

```python
plt.figure(figsize=(10,6))

plt.plot(alist, E_list_IT, 'o')

y_interpolated_it = np.interp(alist, alist, E_list_IT)
plt.plot(alist, y_interpolated_it, label='Imaginary time', color='black')

plt.xlabel("a [nm]")
plt.ylabel("E [eV]")
plt.legend()
plt.show()
```
