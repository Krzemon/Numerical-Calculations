```python
import numpy as np
import matplotlib.pyplot as plt
```

# Zadanie 1

```python
def create_full_graph(N: int):
    graph = np.ones((N, N), dtype=int)
    for i in range(N):
        graph[i, i] = 0
    return graph
```

```python
N = 10
graph_1 = create_full_graph(N)
print(graph_1)
```

```python
plt.imshow(graph_1, cmap='gray')
plt.colorbar()
plt.show()
```

```python
x = np.arange(N)
height = np.zeros(N)
for i in range(N):
    height[np.sum(graph_1[i])] += 1
height = height / np.sum(height)

fig = plt.figure(figsize=(7,4))
plt.bar(x, height)
plt.xlim(0, N)
plt.xlabel('Stopień wierzchołka')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title(f'Graf pełny z N = {N} wierzchołkami')
plt.show()
```

# Zadanie 2

```python
def create_p_full_graph(N: int, p: float):
    graph = []
    for i in range(N):
        v = []
        for j in range(N):
            if i == j:
                continue
            if np.random.rand(1) <= p:
                v.append(j)
        graph.append(v)    
    return graph
```

```python
N = 200
p = 1/3
graph_2 = create_p_full_graph(N, p) # N x k, gdzie k <= N; p = k/N
print(graph_2)
```

```python
x = np.arange(N)
height = np.zeros(N)
for i in range(N):
    height[len(graph_2[i])] += 1
height = height / np.sum(height)

fig = plt.figure(figsize=(12,6))
plt.bar(x, height)
plt.xlim(0, N)
plt.xlabel('Stopień wierzchołka')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title(f'Graf z prawdopodobieństwem połączenia krawędzią p = {round(p,3)} oraz N = {N} wierzchołkami')
plt.show()
```
