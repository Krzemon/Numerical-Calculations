```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic(r, x):
    return r * x * (1-x)
```

### Zadanie 1

```python
x_range = np.linspace(0.1, 0.9, 9)
r = 2
steps = 50
print(x_range)
```

```python
plt.figure(figsize=(7,5), dpi=200)

for x in x_range:
    x_ = x
    track = []
    for i in range(steps):
        track.append(x)
        x = logistic(r, x)
    plt.plot(track, label=f'$x_0 = {x_:.1}$')
    plt.scatter([0], [x_], color='r', s = 3.0)


plt.xlabel('$n$')
plt.ylabel('$x_n$')
plt.title(f'ewolucja czasowa układu')
plt.legend(loc=1, fontsize=8)
plt.savefig('zad_1.png')
plt.show()
```

### Zadanie 2

```python
x = 0.5
r_range = [1, 2, 3, 3.5, 3.55, 3.6]
steps = 100

for r in r_range:
    x_ = x
    track = []
    plt.figure(figsize=(7,5), dpi=200)
    for i in range(steps):
        track.append(x)
        x = logistic(r, x)
        
    plt.plot(track, linewidth=0.4)
    plt.scatter(range(steps), track, color='r', s = 0.4)
    plt.xlabel('$n$')
    plt.ylabel('$x_n$')
    plt.title(f'ewolucja czasowa układu dla $r = {r}$')
    plt.savefig(f'zad_2_r_{r}.png')
    plt.show()
```

### Zadanie 3

```python
steps = 10000
x0 = 0.5
dr = 0.001
r_range = np.arange(1.0, 4.0 + dr, dr)
start_idx = 9000

plt.figure(figsize=(12,7), dpi=300)
r_plot = np.ones(steps - start_idx)

for r in r_range:
    map = []
    x = x0
    for i in range(steps):
        if (i >= start_idx):
            map.append(x)
        x = logistic(r, x)
    plt.scatter(r_plot*r, map, color='k', s = 0.01, edgecolors=None, linewidths=0.0)

plt.xlabel('$r$')
plt.ylabel('$x_n$')
plt.title(f'mapa logistyczna $n \in [{start_idx}, {steps}]$')
plt.savefig('zad_3.png')
plt.show()
```
