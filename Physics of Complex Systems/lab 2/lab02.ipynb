```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.animation as animation
import os
```

```python
N = 60
B = {4,6,7,8}
S = {3,5,6,7,8}
grid = np.zeros((N,N))
neighbours = np.zeros((N,N))
time = 100
```

```python
def step():
    for i in range(N):
        for j in range(N):
            neighbours[i][j] = (grid[(i+1)%N][(j+1)%N] + grid[(i+1)%N][j] + grid[(i+1)%N][j-1] + 
                                grid[i][(j+1)%N] + grid[i][j-1] + 
                                grid[i-1][(j+1)%N] + grid[i-1][j] + grid[i-1][j-1])
            if grid[i][j] == 0 and neighbours[i][j] in B:
                grid[i][j] = 1
                continue
            elif grid[i][j] == 1 and neighbours[i][j] in S:
                continue
            grid[i][j] = 0
```

```python
frames_tab = [0, 1, 2, 5, 10, 50, 100]
sim_number = 3
for sim in range(sim_number):
    os.makedirs(f'sim_{sim+1}', exist_ok=True)
    initial_grid = np.random.randint(0,2,(N,N))
    grid = initial_grid.copy()

    for i in range(101):
        if i in frames_tab:
            fig, ax = plt.subplots(figsize=(6,6))
            ax.imshow(grid, cmap='binary')
            plt.axis('off')
            plt.savefig(f'sim_{sim+1}/frame_{i}.png', bbox_inches='tight')
            plt.close(fig)
        step()

    grid = initial_grid.copy()
    fig, ax = plt.subplots(figsize=(6,6))
    im = ax.imshow(grid, cmap='binary')
    plt.axis('off')
    def animate(i):
        im.set_data(grid)
        ax.set_title(f'step = {i}')
        step()
        return im

    frames = 101
    ani = animation.FuncAnimation(fig, animate, frames=frames, repeat=False, interval=30)

    writer = animation.PillowWriter(fps=10, bitrate=2000)
    fig.subplots_adjust(left=0, right=1, top=1, bottom=0)
    ani.save(f'sim_{sim+1}/simulation.gif', writer=writer)
    plt.close(fig)

```
