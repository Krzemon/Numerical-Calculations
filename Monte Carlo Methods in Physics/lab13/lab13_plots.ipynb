```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_2d_data(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            values = list(map(float, line.split()))
            data.append(values)
    return np.array(data)

def read_histogram_data(file_path):
    bins, histogram, theory = [], [], []
    with open(file_path, 'r') as file:
        for line in file:
            values = list(map(float, line.split()))
            bins.append(values[0])
            histogram.append(values[1])
            theory.append(values[2])
    return bins, histogram, theory

def read_simulation_data(file_path):
    x, density, pressure, temperature, velocity, flux, particle_count = [], [], [], [], [], [], []
    with open(file_path, 'r') as file:
        for line in file:
            values = list(map(float, line.split()))
            x.append(values[0])
            density.append(values[1])
            pressure.append(values[2])
            temperature.append(values[3])
            velocity.append(values[4])
            flux.append(values[5])
            particle_count.append(values[6])
    return x, density, pressure, temperature, velocity, flux, particle_count
```

```python
# Wczytanie danych
velocity, counts, theoretical_counts = read_histogram_data('hist3.dat')

# Tworzenie figury i osi z większym rozmiarem
fig, ax = plt.subplots(figsize=(12, 8))

# Rysowanie danych
ax.scatter(velocity, counts, label='Simulation', color='blue', marker='o')
ax.scatter(velocity, theoretical_counts, label='Theory', color='orange', marker='x')

# Ustawianie etykiet i tytułu
ax.set_xlabel('Velocity (V)', fontsize=14)
ax.set_ylabel('f(v)', fontsize=14)
ax.set_title('Velocity Distribution', fontsize=16)

# Dodanie siatki dla lepszej czytelności
ax.grid(True)

# Dodanie legendy
ax.legend(fontsize=12)

# Wyświetlenie wykresu
plt.show()
```

```python
# Definiowanie iteracji do wyrysowania
iterations = [10, 50, 100, 200]

# Tworzenie figury z większym rozmiarem
plt.figure(figsize=(12, 8))

# Rysowanie danych gęstości dla różnych iteracji
for iteration in iterations:
    x, density, _, _, _, _, _ = read_simulation_data(f'./data/nptv_{iteration}.dat')
    plt.plot(x, density, label=f'Iteration = {iteration}', linewidth=2)

# Ustawianie etykiet i tytułu
plt.xlabel('x', fontsize=14)
plt.xlim(0, 2)
plt.title('Density', fontsize=16)
plt.ylabel(r'$\rho$', fontsize=14)

# Dodanie siatki dla lepszej czytelności
plt.grid(True)

# Dodanie legendy
plt.legend(fontsize=12)

# Wyświetlenie wykresu
plt.show()
```

```python
# Tworzenie figury z większym rozmiarem
plt.figure(figsize=(12, 8))

# Rysowanie danych ciśnienia dla różnych iteracji
for iteration in iterations:
    x, _, pressure, _, _, _, _ = read_simulation_data(f'./data/nptv_{iteration}.dat')
    plt.plot(x, pressure, label=f'Iteration = {iteration}', linewidth=2)

# Ustawianie etykiet i tytułu
plt.xlabel('x', fontsize=14)
plt.xlim(0, 2)
plt.title('Pressure', fontsize=16)
plt.ylabel('p', fontsize=14)

# Dodanie siatki dla lepszej czytelności
plt.grid(True)

# Dodanie legendy
plt.legend(fontsize=12)

# Wyświetlenie wykresu
plt.show()
```

```python
# Tworzenie figury z większym rozmiarem
plt.figure(figsize=(12, 8))

# Rysowanie danych temperatury dla różnych iteracji
for iteration in iterations:
    x, _, _, temperature, _, _, _ = read_simulation_data(f'./data/nptv_{iteration}.dat')
    plt.plot(x, temperature, label=f'Iteration = {iteration}', linewidth=2)

# Ustawianie etykiet i tytułu
plt.xlabel('x', fontsize=14)
plt.xlim(0, 2)
plt.title('Temperature', fontsize=16)
plt.ylabel('T', fontsize=14)

# Dodanie siatki dla lepszej czytelności
plt.grid(True)

# Dodanie legendy
plt.legend(fontsize=12)

# Wyświetlenie wykresu
plt.show()
```
