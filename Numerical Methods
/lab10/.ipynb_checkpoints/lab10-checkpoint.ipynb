```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(x, y):
    return 5/2 * (x**2 - y)**2 + (1 - x)**2

def deriv_x_vec(f, x, y, delta=1e-4):
    return (f(x + delta, y) - f(x - delta, y)) / (2*delta)

def deriv_y_vec(f, x, y, delta=1e-4):
    return (f(x, y + delta) - f(x, y - delta)) / (2*delta)
```

```python
r_vec = np.zeros(2)
r_vec_next = np.zeros(2)
r_vec[0] = -0.75
r_vec[1] = 1.75
h = 0.1
VEC_A = []

eps_a = 1e-2
counter_a = 0

while(True and counter_a < 1000):
    counter_a += 1
    r_vec_next = r_vec - h * np.array([deriv_x_vec(f, r_vec[0], r_vec[1]), deriv_y_vec(f, r_vec[0], r_vec[1])])
    if np.sqrt((r_vec_next[0] - r_vec[0])**2 + (r_vec_next[1] - r_vec[1])**2) < eps_a:
        break
    r_vec = r_vec_next
    VEC_A.append(r_vec)

print("Wynik: ", r_vec)
print("Liczba kroków: ", counter_a)
```

```python
VEC_A
```

```python
X_A = np.transpose(VEC_A)[0]
Y_A = np.transpose(VEC_A)[1]
```

```python
r_vec = np.zeros(2)
r_vec_next = np.zeros(2)
r_vec[0] = -0.75
r_vec[1] = 1.75
h = 0.1
VEC_B = []

eps_b = 1e-3
counter_b = 0


while(True and counter_b < 1000):
    counter_b += 1
    r_vec_next = r_vec - h * np.array([deriv_x_vec(f, r_vec[0], r_vec[1]), deriv_y_vec(f, r_vec[0], r_vec[1])])
    if np.sqrt((r_vec_next[0] - r_vec[0])**2 + (r_vec_next[1] - r_vec[1])**2) < eps_b:
        break
    r_vec = r_vec_next
    VEC_B.append(r_vec)

print("Wynik: ", r_vec)
print("Liczba kroków: ", counter_b)
```

```python
VEC_B
```

```python
X_B = np.transpose(VEC_B)[0]
Y_B = np.transpose(VEC_B)[1]
```
