```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def F_base(x):
    return np.log(x) * (x - 2)**2

def F_d1(x):
    return (x - 2) * (1 -  2/x + 2 * np.log(x))

def F_d2(x):
    return 3 -  4/x + 2*np.log(x) - 4 / (x**2)
```

# Zadanie 1

```python
N = 100

X = np.linspace(0.5, 2.4, N)
F = np.zeros(N)

for i,x in enumerate(X):
    F[i] = F_base(x)
    
plt.figure(figsize=(10,6))
plt.plot(X, F)

plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('F(x) = ln(x) * (x - 2)**2')
plt.grid(True)
plt.show()
```

```python
Fx = np.zeros(N)

for i,x in enumerate(X):
    Fx[i] = F_d1(x)
    
plt.figure(figsize=(10,6))
plt.plot(X, Fx)

plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('F\'(x) = (x - 2) * (1 -  2/x + 2ln(x))')
plt.grid(True)
plt.show()
```

```python
Fxx = np.zeros(N)

for i,x in enumerate(X):
    Fxx[i] = F_d2(x)
    
plt.figure(figsize=(10,6))
plt.plot(X, Fxx)

plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('F\'\'(x) = 3 -  4/x + 2ln(x) - 4 / (x**2)')
plt.grid(True)
plt.show()
```

# Zadanie 3

```python
zeros = []
eps = 1
x0 = 2.3
n = 1
zeros.append(x0)

while True:
    x_curr = zeros[n - 1]
    x_next = x_curr - F_base(x_curr) / F_d1(x_curr)
    zeros.append(x_next)
    eps = x_next - x_curr
    n += 1     # aktualna liczba przyblizen miejsca zerowego
    if abs(eps) < 1e-6:
        break
```

```python
num = np.arange(n)

    
plt.figure(figsize=(10,6))
plt.plot(num, zeros)

plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('przyblizenia miejsca zerowego w poblizu x = 2')
plt.grid(True)
plt.show()
```

# Zadanie 5

```python
def U_base(x):
    return F_base(x) / F_d1(x)

def U_d1(x):
    return 1 - F_d2(x) / F_d1(x) * U_base(x)
```

# A

```python
zeros = []
eps = 1
x0 = 2.3
n = 1
zeros.append(x0)

while True:
    x_curr = zeros[n - 1]
    x_next = x_curr - 2 * F_base(x_curr) / F_d1(x_curr)
    zeros.append(x_next)
    eps = x_next - x_curr
    n += 1     # aktualna liczba przyblizen miejsca zerowego
    if abs(eps) < 1e-6:
        break
```

```python
num = np.arange(n)

    
plt.figure(figsize=(10,6))
plt.plot(num, zeros)

plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('przyblizenia miejsca zerowego w poblizu x = 2')
plt.grid(True)
plt.show()
```
