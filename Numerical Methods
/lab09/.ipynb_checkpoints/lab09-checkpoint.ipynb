```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def fun_1(x, alfa = 0):
    return 2*np.sin(x) + np.sin(2*x) + 2*np.sin(3*x) + alfa

def fun_2(x):
    return 2*np.sin(x) + np.sin(2*x) + 2*np.cos(x) + np.cos(2*x)

def fun_3(x):
    return 2*np.sin(1.1*x) + np.sin(2.1*x) + 2*np.sin(3.1*x)
```

```python
# Funkcja aproksymująca
def F(x, a, b):
    Ns = len(a)
    Nc = len(b)
    result = np.zeros_like(x)
    for j in range(Ns):
        result += a[j] * np.sin(j * x)
    for j in range(Nc):
        result += b[j] * np.cos(j * x)
    return result
```

```python
# Parametry
n = 100
Ns = 5
Nc = 5
X = np.linspace(0, 2*np.pi, n)
dx = 2*np.pi / n

# Generowanie danych funkcji f1(x)
f1 = fun_1(X)

# Wyznaczanie współczynników ak i bj
a1 = np.zeros(Ns)
b1 = np.zeros(Nc)

for j in range(Ns):
    a1[j] = np.sum(f1 * np.sin(j * X)) / n

for j in range(1, Nc):
    b1[j] = np.sum(f1 * np.cos(j * X)) / n

b1[0] = np.sum(f1) / (2 * n)
```

## Zadanie 1

```python
# Wykres funkcji f1(x) i F1(x)
plt.figure(figsize=(12, 7))

plt.plot(X, f1, label='f1(x)', color='r', linestyle='--')
plt.plot(X, F(X, a1, b1), label='F1(x)', color='g', linestyle='-')

plt.xlabel('X')
plt.ylabel('$f_{1}(x)$')
# plt.title('Aproksymacja funkcji f1')

plt.legend()
plt.grid(True)
plt.show()

print("Współczynniki ak:")
print([round(coeff, 5) for coeff in a1])
print("Współczynniki bj:")
print([round(coeff, 5) for coeff in b1])
```

```python
# Funkcja generująca wartości α zgodnie z podanym wzorem
def generate_alpha():
    return np.random.uniform(-0.5, 0.5)

# Parametry
n = 100
X = np.linspace(0, 2*np.pi, n)
dx = 2*np.pi / n

# Generowanie danych funkcji f1(x)
f1 = np.zeros(n)
for i in range(n):
    f1[i] = fun_1(X[i], alfa=generate_alpha())

# Zestawy parametrów Ms i Mc
params_sets = [(5, 5), (30, 30)]

for Ms, Mc in params_sets:
    # Wyznaczanie współczynników ak i bj dla f1(x)
    a1 = np.zeros(Ms)
    b1 = np.zeros(Mc)

    for j in range(Ms):
        a1[j] = np.sum(f1 * np.sin(j * X)) / n

    for j in range(1, Mc):
        b1[j] = np.sum(f1 * np.cos(j * X)) / n

    b1[0] = np.sum(f1) / (2 * n)

    # Wykres funkcji f1(x) i F1(x)
    plt.figure(figsize=(12, 7))

    plt.plot(X, f1, label='f1(x)', color='r', linestyle='--')
    plt.plot(X, F(X, a1, b1), label='F1(x)', color='g', linestyle='-')

    plt.xlabel('X')
    plt.ylabel('$f_{1}(x)$')
#     plt.title('Aproksymacja funkcji f1 (Ms={}, Mc={})'.format(Ms, Mc))

    plt.legend()
    plt.grid(True)
    plt.show()

    # Wykres wartości współczynników ak i bj
    plt.figure(figsize=(12, 7))

    plt.plot(range(Ms), a1, label='ak', color='b', marker='o')
    plt.plot(range(Mc), b1, label='bj', color='m', marker='s')

    plt.xlabel('Index')
    plt.ylabel('Value')
#     plt.title('Wartości współczynników ak i bj (Ms={}, Mc={})'.format(Ms, Mc))

    plt.legend()
    plt.grid(True)
    plt.show()
```
