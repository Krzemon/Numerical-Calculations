```python
import numpy as np
from numpy import sin, sqrt, pi, exp
from matplotlib import pyplot as plt
from random import random
from scipy.fft import fft, ifft
```

```python
T = 1 # okres trwania sygnalu
k_tab = [8, 10, 12]
σ = T / 20
t_max = 3 * T
ω = 2 * pi / T

def f_base(t):
    return sin(1*ω*t) + sin(2*ω*t) + sin(3*ω*t)

def g_base(t):
    return 1 / (sqrt(2*pi)) * exp(-t*t / (2*σ*σ))
```

a) f0[ ] dla sygnału bez szumu (wzór 5)

b) f [ ] dla sygnału z szumem (wzór 4)

c) dwie tablice dla funkcji wagowej: g1 i g2 - do obu wpisujemy te same wartości (wzór 6).

```python
# funkcja zwracajaca najwiekszy modul liczby zespolonej

def complex_max(a):
    length = np.size(a)
    mam = np.abs(a[0])
    for i in range(1, length):
        if mam < np.abs(a[i]):
            mam = np.abs(a[i])
    return mam
```

```python
k = k_tab[0]
#########################################
N_k = np.power(2, k) # liczba wezlow 
dt = t_max / N_k
f0 = np.zeros(2 * N_k, dtype = complex)
f = np.zeros(2 * N_k, dtype = complex)
g1 = np.zeros(2 * N_k, dtype = complex)
g2 = np.zeros(2 * N_k, dtype = complex)

time = np.linspace(0, 2 * N_k, 2*N_k)

for i in range(0, 2*N_k):
    f0[i] = f_base(i * dt)
    f[i] = f0[i] + random() - 0.5 # zaszumiamy
    g1[i] = g_base(i * dt)
    g2[i] = g_base(i * dt)

fk = fft(f)
g1k = fft(g1)
g2k = ifft(g2)

for i in range(0, 2*N_k):
    f[i] = fk[i] * (g1k[i] + g2k[i])

splot = ifft(f)
```

```python
max_value = complex_max(splot)
```

```python
print(max_value)
```

```python
splot_norm = splot * 2.5 / max_value
```

```python
plt.figure(figsize = (13,7))

plt.plot(time, f, label='f1(x)', color='g', linestyle='--')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.ylim(-10, 10)
plt.show()
```

```python
plt.figure(figsize = (13,7))

plt.plot(time, splot_norm, label='f1(x)', color='r', linestyle='--')

plt.xlabel('time')
plt.ylabel('values')

plt.show()
```

```python
plt.figure(figsize = (13,7))

plt.plot(time, splot_norm, label='f1(x)', color='r', linestyle='--')
plt.plot(time, f0, label='f1(x)', color='g', linestyle='--')

plt.xlabel('time')
plt.ylabel('values')

plt.show()
```
